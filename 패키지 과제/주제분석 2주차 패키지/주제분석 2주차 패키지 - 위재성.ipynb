### 3. 불필요한 변수 제거

```python
# 고유값 하나뿐인 열 제거
data = data.drop(columns=['Z_CostContact','Z_Revenue'])

# 고유값 개수가 열의 개수와 같은 열 제거
data = data.drop(columns=['ID'])

data.nunique()
```

```python
len(data)
```

### 4. 결측치 처리

```python
data['Income'] = data['Income'].fillna(np.mean(data.Income))
```

```python
data.isna().sum()
```

### 5. 재범주화

```python
for level in data['Marital_Status'].unique():
    if level == 'Together' or level == 'Married':
        data['Marital_Status'][data['Marital_Status'] == level] = 'Relationship'
    else:
        data['Marital_Status'][data['Marital_Status'] == level] = 'Single'
```

```python
relate_len = len(data['Marital_Status'][data['Marital_Status'] == 'Relationship'])
single_len = len(data['Marital_Status'][data['Marital_Status'] == 'Single'])
print('Relationship:',relate_len ,' Single:',single_len)
```

```python
for level in data['Education'].unique():
    if level == 'Basic' or level == '2n Cycle':
        data['Education'][data['Education'] == level] = 'Undergraduate'
        
    elif level == 'Graduation':
        data['Education'][data['Education'] == level] = 'Graduate'
        
    else:
        data['Education'][data['Education'] == level] = 'Postgraduate'
```

```python
under_len = len(data['Education'][data['Education'] == 'Undergraduate'])
grad_len = len(data['Education'][data['Education'] == 'Graduate'])
post_len = len(data['Education'][data['Education'] == 'Postgraduate'])
print('Undergraduate:',under_len ,' Graduate:',grad_len, ' Postgraduate:',post_len)
```

## Chapter2. 파생변수 생성

### 1. 가정 내 어린이 수 및 부모 여부

```python
data['Kids'] = data.Kidhome + data.Teenhome

data['Is_Parent'] = 0
data.loc[data['Kids'] > 0, 'Is_Parent'] = 1
```

```python
data[['Kidhome','Teenhome','Kids','Is_Parent']]
```

### 2. 나이

```python
data['Age'] = 2021 - data['Year_Birth']
```

```python
data[['Age','Year_Birth']]
```

### 3. 등록일로부터 일수

```python
data['Today'] = pd.to_datetime('2021-11-02') # 현재날짜
data['Day_engaged'] = data['Today'] - data['Dt_Customer'].apply(lambda x: pd.to_datetime(x))
```

```python
# days 문자열 삭제
data['Day_engaged'] = data['Day_engaged'].astype('string').str.replace('days','')
data[['Day_engaged']]
```

### 4. 총 소비

```python
data['Expenses'] = data[['Wines','Fruits','MeatProducts','FishProducts','SweetProducts','GoldProds']].sum(axis=1)
```

```python
data[['Wines','MeatProducts','FishProducts','SweetProducts','GoldProds','Expenses']]
```

### 5. 불필요한 변수 삭제

```python
drop_list = ['Kidhome','Teenhome','Year_Birth','Dt_Customer','Today']
data = data.drop(columns = drop_list)
```

## Chapter3. 시각화

### 1. Bar Graph

```python
sns.set_theme()
sns.set_context('talk')
sns.set_style('ticks')
fig, axs = plt.subplots(ncols=2, figsize=(18,10))

ax1 = sns.countplot(data['Education'], ax = axs[0], palette='Set2')
ax2 = sns.countplot(data['Marital_Status'], ax = axs[1], palette='husl')

ax1.set_title('Education Level of Customers')
ax2.set_title('Marital Status of Customers')
```

### 2. Histogram & Boxplot

```python
fig, axs = plt.subplots(ncols=2, figsize=(20,13))

ax1 = sns.distplot(data['Age'], ax=axs[0], kde=False, 
                   hist_kws=dict(edgecolor="black", linewidth=2, alpha=0.8))
ax2 = sns.boxplot(y=data['Age'], ax=axs[1])

ax1.set_title('Ages of Customers')
ax2.set_title('Boxplot of Ages')

ax1.set_ylabel('')
ax1.xaxis.grid(True)
```

```python
data['Age'].sort_values(ascending=False)
```

```python
data = data.loc[data['Age'] < 120,:] # 120 미만으로만 저장
```

```python
fig, axs = plt.subplots(ncols=2, figsize=(20,13))
ax1 = sns.distplot(data['Age'], ax=axs[0], kde=False, color='pink',
                   hist_kws=dict(edgecolor="black", linewidth=2, alpha=0.8))
ax2 = sns.boxplot(y=data['Age'], ax=axs[1], color='pink')

ax1.set_title('Ages of Customers', fontsize=20)
ax2.set_title('Boxplot of Ages', fontsize=20)

ax1.set_ylabel('')
ax1.xaxis.grid(True)
```

### 2-2. Histogram

```python
var = {'Wines':'darkorange',
 'Fruits':'seagreen',
 'MeatProducts':'lightcoral',
 'SweetProducts':'mediumturquoise'}
```

```python
fig, axs = plt.subplots(figsize=(15,13))
fig.suptitle('Items bought by Customers', fontsize = 30)

for i,(name,color) in enumerate(var.items()):
    plt.subplot(2,2,i+1)
    ax = sns.distplot(data[name], color=color,kde=False,
                  hist_kws=dict(edgecolor="black", linewidth=1, alpha=0.8))
    ax.set_ylabel('')
    ax.xaxis.grid(True)
    
    if i==0:
        ax.set_xlabel('Wine')
    elif i==1:
        ax.set_xlabel('Fruit')
    elif i==2: 
        ax.set_xlabel('Meat')
    else:
        ax.set_xlabel('Sweets')
```

### 3. Scatterplot

```python
fig, ax = plt.subplots(figsize = (15,15))

ax = sns.scatterplot(data=data, x='Income', y='Wines', hue='Is_Parent')

plt.title('Income vs. Wine Spending', fontsize=20)
ax.set_xlim(0,200000)
ax.set_xlabel('Income [USD $]', fontsize=20)
ax.set_ylabel('Wine Spending [USD $]', fontsize=20)

# 범례 수정
h, l = ax.get_legend_handles_labels()
ax.legend(h, ['Has Child', 'No Child'], title='' )
```
